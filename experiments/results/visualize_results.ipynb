```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('splm_mnist_hinge_2.csv', index_col=0)
```

```python
df.groupby('batch_size').max().sort_values('best_eval_accuracy', ascending=False)
```

```python
df[df['status'] == 'successful'].groupby('batch_size').max()['best_eval_accuracy'].plot()
plt.title('eval accuracy vs. batch_size')
plt.ylabel('best eval accuracy')
plt.xlabel('batch_size')
plt.show()
```

```python
df[df['status'] == 'successful'].groupby('batch_size').mean()['mean_epoch_time'].plot()
plt.title('epoch time vs. batch_size')
plt.ylabel('epoch time (seconds)')
plt.xlabel('batch_size')
plt.show()
```

```python
df[df['status'] == 'successful'].groupby('beta').max()['best_eval_accuracy'].plot(logx=True)
plt.title('best eval accuracy vs. beta')
plt.ylabel('eval accuracy')
plt.xlabel('beta (log)')
plt.show()
```

```python
df = pd.read_csv('splm_mnist_hinge_1.csv', index_col=0)
```

```python
df.groupby('K').mean()['mean_epoch_time'].plot()
plt.xlabel('K (inner iterations)')
plt.ylabel('mean epoch time (seconds)')
plt.title('epoch time vs. K')
plt.ylim((0, 23))
plt.xlim((0, 100))
plt.show()
```
