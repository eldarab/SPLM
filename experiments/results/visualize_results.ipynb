# scheduler

```python
df = pd.read_csv('splm_mnist_hinge_scheduler_results.csv', index_col=0)
```

```python
df[df['status'] == 'successful']['best_train_accuracy'].plot.hist()
plt.xlabel('best_train_accuracy')
plt.title('best_train_accuracy histogram')
plt.show()
```

```python
df[df['status'] == 'successful'].sort_values('best_train_accuracy', ascending=False)
```

```python
df.groupby('status').describe()['step_size']#.sort_values('best_train_accuracy', ascending=False)
```

```python
def plot_hparam_effect(hparam):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    ax1.plot(df[df['status'] == 'successful'].groupby(hparam).mean()['best_eval_accuracy'])
    ax1.plot(df[df['status'] == 'successful'].groupby(hparam).mean()['best_train_accuracy'])
    ax1.set_title(f'best accuracy vs. {hparam}')
    ax1.set_ylabel('accuracy')
    ax1.set_xlabel(hparam)
    ax1.legend(['best_eval_accuracy', 'best_train_accuracy'])
    ax1.set_ylim(0, 1)
    
    eval_loss = df[df['status'] == 'successful'].groupby(hparam).mean()['best_eval_loss']
    train_loss = df[df['status'] == 'successful'].groupby(hparam).mean()['best_train_loss']
    ax2.plot(eval_loss)
    ax2.plot(train_loss)
    ax2.set_title(f'best loss vs. {hparam}')
    ax2.set_ylabel('loss')
    ax2.set_xlabel(hparam)
    ax2.legend(['best_eval_loss', 'best_train_loss'])
    ax2.set_ylim(0, 1.1 * max(list(eval_loss) + list(train_loss)))
    
    plt.show()
```

```python
plot_hparam_effect('beta')
```

```python
plot_hparam_effect('gamma')
```

```python
plot_hparam_effect('step_size')
```
