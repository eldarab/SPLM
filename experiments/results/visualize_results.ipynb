```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('splm_mnist_hinge_2.csv', index_col=0)
```

```python
df[df['status'] == 'successful'][df['beta'] == 316].groupby('batch_size').mean()['best_train_accuracy'].plot()
```

```python
df[df['beta'] == 100]
```

```python
df.groupby(['status', 'beta']).mean()
```

```python
plot = df[df['status'] == 'successful'].groupby(['beta']).mean()
plt.scatter(plot.index, plot.best_eval_accuracy)
plt.ylabel('eval accuracy')
plt.xlabel('beta')
plt.title('accuracy vs beta')
plt.show()
```
