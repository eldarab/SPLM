```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results_1.csv', index_col=0)
```

```python
df[df['status'] == 'failed'][df['batch_size'] == 2]
```

```python
df = df[df['status'] == 'successful']
```

```python
df.groupby(['batch']).mean().sort_values('mean_epoch_time')
```

```python
df.groupby('batch_size').mean().sort_values('best_eval_accuracy', ascending=False)
```

```python
df.groupby('K').mean().sort_values('best_eval_accuracy', ascending=False)
```

```python
plt.scatter(df['beta'], df['best_train_accuracy'])
plt.ylabel('accuracy')
plt.xlabel('beta')
plt.title('accuracy vs beta')
plt.show()
```
